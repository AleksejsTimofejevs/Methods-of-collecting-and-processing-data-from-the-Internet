```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
from datetime import datetime as dt

%matplotlib inline
```

```python
rcParams["figure.figsize"] = 15, 8
```

```python
df = pd.read_csv('/Users/a1/Desktop/Geek/Methods-of-collecting/Lesson-08/opendata.csv', encoding='windows-1251')
```

```python
def open_data(name, region, start_date, end_date):
    df_new = df[(df['name']==name) & (df['region']==region) & df['date'].between(start_date, end_date)]
    rcParams["figure.figsize"] = 15, 8
    plt.plot(pd.to_datetime(df_new['date']).dt.strftime('%m:%y'), df_new['value'], label=region)
    plt.scatter(pd.to_datetime(df_new['date']).dt.strftime('%m:%y'), df_new['value'])
    plt.title(name)
    legend = plt.legend(loc="lower right")
    plt.grid()
    plt.show()
```

```python
name = 'Средняя сумма заявки на ипотечный кредит'
region = 'Россия'
start_date = '2014-01-15'
end_date = '2016-01-15'
```

```python
open_data(name, region, start_date, end_date)
```
